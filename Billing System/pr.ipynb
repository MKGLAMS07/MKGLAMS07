```python
import tkinter as tk

class BillingApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Billing App")

        self.product_label = tk.Label(master, text="Product:")
        self.product_label.pack()

        self.product_entry = tk.Entry(master)
        self.product_entry.pack()

        self.price_label = tk.Label(master, text="Price:")
        self.price_label.pack()

        self.price_entry = tk.Entry(master)
        self.price_entry.pack()

        self.add_button = tk.Button(master, text="Add Item", command=self.add_item)
        self.add_button.pack()

        self.total_label = tk.Label(master, text="Total: $0.00")
        self.total_label.pack()

        self.reset_button = tk.Button(master, text="Reset", command=self.reset)
        self.reset_button.pack()

        self.items = []
        self.total = 0.0

    def add_item(self):
        product = self.product_entry.get()
        price = float(self.price_entry.get())

        self.items.append((product, price))
        self.total += price

        self.update_total()
        self.reset_entries()

    def update_total(self):
        self.total_label.config(text=f"Total: ${self.total:.2f}")

    def reset_entries(self):
        self.product_entry.delete(0, tk.END)
        self.price_entry.delete(0, tk.END)

    def reset(self):
        self.items = []
        self.total = 0.0
        self.update_total()

if __name__ == "__main__":
    root = tk.Tk()
    app = BillingApp(root)
    root.mainloop()
```
